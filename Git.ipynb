<h1 style="color:black" align="center">1. Подготовка</h1>

### Установка имени и электронной почты

`git config --global user.name "Your Name"` - Установка имени

`git config --global user.email "your_email@whatever.com"` - Установка почты

### Параметры установки окончаний строк

`git config --global core.autocrlf true`

`git config --global core.safecrlf warn`

### Установка отображения unicode

По умолчанию, git будет печатать не-ASCII символов в именах файлов в виде восьмеричных последовательностей \nnn. Что бы избежать нечитаемых строк, установите соответствующий флаг.

`git config --global core.quotepath off`

<h1 style="color:black" align="center">2. Работа с репозиториями в Git</h1>

### Создание репозитория с синхронизацией

Для создания репозитория требуется перейти в нужную директорию и выполнить команду:

`git init` - Создание пустого репозитория в текущем местоположении с добавлением своей собственной папки. Можно проверить командой `ls -a`

`git status` - Используется чтобы проверить текущее состояние репозитория.

`git remote add origin "link to repository"` - Создание синхронизация удалённого сервера с локальными файлами

`git remote -v` - Проверка синхронизации

<h1 style="color:black" align="center">3. Управление файлами репозитория</h1>

### Создание комита

Изначально требуется иметь изменения либо какой-то файл для создания commit. Предположим, что данный файл у нас уже есть, назовём его test.txt. 

Используем команду: 

`git status` - В данной ситуации она покажет, текущее состояние репозитория, позволит определить наличие изменений для созданий commit.

`git restore "file name"` - Отменение изменений файла, на прошлое состояние.

После выполнения команды проверки статуса, test.txt будет находиться в категории **Untracked files:** - Это файлы, которые были добавлены в репозиторий, но ещё не зарегестрированы в индексе изменений. Для этого требуется добавить файл в индекс:

`git add "file name"` - Регистрация файла в индексе изменений. 

После выполнения команды регистрации файла, он будет отправлен в категорию **Changes to be committed** - Категория, которая отвечает за изменения, которые будут записаны при создании commit.

Если была произведена ошибка, то его можно удалить из спика файлов, которые отправляютс в commit:

`git rm --cached "file name"` - Удаление файла из списка commit.

На данном этапе, можно выполнить commit:

`git commit` - Создание commit, после чего запускается редактор vim, который установлен по умолчанию для создания сообщения, которое будет привязано commit.

`git commit -m "commit message"` - Записать commit с сообщением, без открытия редактора.

**Краткая версия commit:**

`git commit -a` - Является сочетанием сразу двух команд git add + git commit. Команду можно ещё модифицировать, добавив флаг `-am`, что позволит произвести commit с сообщением сразу.

**Служебные файлы:**
 
`.gitkeep` - Позволяет сохронять пустые директории, которые в будущем могу быть заполенены какими-то файлами. При создание директории она будет не определена, для этого её требуется занести в данный файл командой `touch images/.gitkeep`. `images/` or `storage/`.

`.gitignore` - Игнорирование некоторых файлов, от попадания их в общий репозиторий. `.env` or `temp/*` or `cache/*`

<h1 style="color:black" align="center">4. История</h1>

`git log` - Предоставляет информацию о каждом commit, напротив которого написан его хэш, автор, время создания и его комментарий.

`git log --oneline` - Вывод краткой истории

`git reset  хэш` - Возврат к прошлому commit, хэш которого был указан

`git reflog` - История всех операций, которые были совершены во время работы с репозиторием (отмена commit, переход по истории...). Благодоря этому, после выполнения команды reset, можно найти хэш потерянного commit и вернутся на него, выполняется так же командой `git reset  хэш`.

`git revert хэш` - Отличается от reset тем, что она не сбрасывает состояние нашего репозитория на конкретный commit, а создаёт новый commit, который отменят дейсвия одно из commit, который был выполнен ранее.

<h1 style="color:black" align="center">5. Ветки</h1>

`git branch` - Показывает доступные ветки.

`git branch "new branch name"` - Создание ветки.

`git checkout "new branch name"` - Переход на другую ветку.

`git checkout "new branch name" "old branch name"`- Создание ветки с указанием, он какой будет она начинатсья (точка старта).

`git checkout -b "branch name"` - Создание ветки с переключением на неё, если такой ветки нет, то она будет создана автоматически, за это отвечает флаг -b.

<h1 style="color:black" align="center">6. Получение удалённого репозитория через SSH (clone)</h1>

**Клонирование репозитория:**

Для настройки SSH-подключения необходимо создать ключ, для этого необходимо выполнить команду:

`ssh-keygen` - Генерация уникального ключа, который вноситься в настройки GitHub и автоматически передаётся при каждой загрузки или скачивания данных с репозитория. Данная команда изначально просит указать путь для сохранения файла, после кодовое слово, которое необязательно вводить, можно просто нажать Entr, если ключ не выбран, тогда необходимо нажать два раза.

После этого мы переходим в аккаунт GitHub -> sttings -> SSH and GPG keys -> new SSH key -> указываем название -> в большое поле вставляем содержимое ключа с файла, который был сгенерирован. После чего будет создан ключ, где будет указана основаня информация, когда был добавлен и был ли он вообще использован.

После переходим в репозиторий, который хотим использовать, клонируем репозиторий выбирая тип подключения SSH и переходим в терменал и выполняем команду:

`git clone "путь" "папка, куда будет клонироваться репозиторий"`

remote будет установлен автоматически, это можно проверить командой:

`git remote -v`

**Загрузка изменений на удалённый репозиторий**

`git push` - Загружает все изменения в удалённый репозиторий.

В случае, когда мы создали новую ветку на локальном репозитории, тогда необходимо создать её и на удалённом тоже и только тогда можно будет выполнить комаду push.

`git push --set-upstream origin dev` - Создание ветки в автоматическом режиме. Где --set-upstream origin - это источник в удалённом репозитории под названием origin. origin - это remote-репозиторий. 

Краткая версия данной команды

`git push -u origin dev` - Загрузка с текущей ветки dev, на ветку origin dev.

**Получение изменений:**

`git pull` - Загрузка последних обновлений ветки.

`git fetch` - Обновление данных из репозитория (получение новой ветки...).
