```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Настройка
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

### Создаём словари, для визуализации текста на графике

```python
title_dict = {'fontsize': '16', 'fontweight': 'bold', 'color': '#808080', 'family': 'Calibri'}
lable_dict = {'fontsize': '12', 'color': '#808080', 'family': 'Calibri'}
```

### Создаём данные

```python
x = np.arange(1, 8)
x
```

```python
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
y
```

### Строим графики

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax

# Линейный график
ax1.plot(x, y)
ax1.set_title('Линейный график', fontdict=title_dict)
ax1.set_xlabel('Ось X', fontdict=lable_dict)
ax1.set_ylabel('Ось Y', fontdict=lable_dict)
 
# Диаграмма разброса
ax2.scatter(x, y)
ax2.set_title('Диаграмма разброса', fontdict=title_dict)
ax2.set_xlabel('Ось X', fontdict=lable_dict)
ax2.set_ylabel('Ось Y', fontdict=lable_dict)

# Настройка фигур
fig.set_size_inches(10, 5)
plt.subplots_adjust(wspace=0.3)
```

### Подготовка среды для работы

```python
import pandas as pd
```

```python
# Стиль графика
plt.style.use('fivethirtyeight')
```

### Чтение DataSet

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.shape
```

### Подсчёт количества наблюдений

```python
class_count = df['Class'].value_counts()
class_count
```

### Построение столбчатой диаграммы

```python
class_count.plot(kind='bar')

plt.title('Линейный масштаб', fontdict=title_dict)
plt.xlabel('Class', fontdict=lable_dict)
plt.ylabel('Представления', fontdict=lable_dict)

plt.show()
```

### Построение столбчатой диаграммы с логарифмическим масштабом

```python
class_count.plot(kind='bar')

plt.yscale(value='log')

plt.title('Логарифмический масштаб', fontdict=title_dict)
plt.xlabel('Class', fontdict=lable_dict)
plt.ylabel('Представления', fontdict=lable_dict)

plt.show()
```

### Class = 1 and Class = 0

```python
class_count
```

```python
class_0 = df.loc[df['Class'] == 0, 'V1']
class_1 = df.loc[df['Class'] == 1, 'V1']
```

```python
# График - мошеннические транзакции
plt.hist(class_1, bins=20, density=True, alpha=0.5, label='Class 1', color='red')

# График - обычные транзакции
plt.hist(class_0, bins=20, density=True, alpha=0.5, label='Class 0', color='grey')

plt.title('Транзакции', fontdict=title_dict)
plt.xlabel('Class', fontdict=lable_dict)
plt.ylabel('Плотность распределения', fontdict=lable_dict)

plt.legend(fontsize=12, frameon=False)

plt.show()
```
