```python
A = set([10, 20, 30])
B = set([20, 40, 50])
C = set([20, 40, 60])
```

**Объединение**

```python
A.union(B, C)
```

**Пересечение**

```python
A.intersection(C)
```

**Разность**

```python
C.difference(A)
```

**Симметрическая разность**

```python
A.symmetric_difference(B)
```
