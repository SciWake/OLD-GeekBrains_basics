### Задание 
Напишите уравнение параболы, проходящей через три точки (x,y):
(1,2), (3,10), (5,1)

__Решение:__

Парабола задается уравнением


$y = ax^2 + bx + c$, где

$a$, $b$ и $c$ — коэффициенты параболы, которые нам требуется найти.

Подставляем в это уравнение заданные координаты точек и получаем систему:

$\begin{cases}
a + b + c = 2, \\
9a + 3b + c = 10 \\
25a + 5b + c = 1
\end{cases}$

Пользуясь методом сложения, упрощаем систему. Умножим первое уравнение на -1 и складываем с остальными поочерёдно и получаем систему вида:

$\begin{cases}
8a + 2b = 8 \\
24a + 4b = -1
\end{cases}$

Решаем уравнение $8a + 2b = 8$ относительно уравнения $24a + 4b = -1$:

$\begin{matrix}
8a + 2b = 8 \\
2b = 8 - 8a \\
b = 4 - 4a
\end{matrix} \Rightarrow$
$\begin{matrix}
24a + 4b = -1 \\
24a + 16 - 16a = -1 \\
8a = -17 \\
a = - \frac{17}{8}
\end{matrix}$

Подставляем $a$ в уравнение $b = 4 - 4a$:

$b = 4 - 4a$

$b = 4 + \frac{17}{2}$

$b = \frac{25}{2}$

Находим переменную c, подставив одно из значений в уравнение $a + b + c = 2$:

$a + b + c = 2$

$c = 2 + \frac{17}{8} - \frac{25}{2}$

$c = \frac{33}{8} - \frac{100}{8}$

$c = -\frac{67}{8}$

**Корин системы:**
$(a, b, c) = (-\frac{17}{8}, \frac{25}{2}, -\frac{67}{8})$

**Ответ:**
$y = -\frac{17}{8}x^2 + \frac{25}{2}x -\frac{67}{8}$
