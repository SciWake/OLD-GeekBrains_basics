# **Первая часть**

__1.__ Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

Решение:

$$\lambda_{1}f_{1}+\lambda_{2}f_{2}+\lambda_{3}f_{3}+\lambda_{4}f_{4} = 0$$
$$\lambda_{1}*e^{x}+\lambda_{2}*1+\lambda_{3}*(x+1)+\lambda_{4}*(x-e^{x}) = 0$$

Возьмем $\lambda_{4} = \lambda_{1} = - \lambda_{3} = \lambda_{2} = 1$. Тогда

$$e^{x}+1-(x+1)+(x-e^{x}) = 0$$

Получается, что система линейно зависима.

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}.$$

Решение:

$$\lambda_{1}f_{1}+\lambda_{2}f_{2}+\lambda_{3}f_{3}+\lambda_{4}f_{4} = 0$$
$$\lambda_{1}*2+\lambda_{2}*x+\lambda_{3}*x^{2}+\lambda_{4}*(x+1)^{2} = 0$$

Возьмем $\lambda_{4} = - \lambda_{3} = 2,  \lambda_{2} = 2 * \lambda_{4} = 4, \lambda_{1} = 1/2 * \lambda_{4} = 1$. Тогда

$$2+4*x+2*x^{2}+2*(x+1)^{2} = 0$$

Получается, что система линейно зависима.

$$x = 1/2*b_1 + 1*b_2 + 3*b_3 $$

$||x_2|| = ||(12,3,4)|| = \sqrt{12^2+3^2+4^2} = \sqrt{169} = 13$

$$cos\varphi = \frac{(x_1,x_2)}{\left \|x_1\right \|\cdot\left \|x_2\right \|} = \frac{4*12+2*3+4*4}{6*13} = \frac{70}{78} = \frac{35}{39}.$$
