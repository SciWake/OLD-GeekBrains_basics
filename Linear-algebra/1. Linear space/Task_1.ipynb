Найти нормы векторов  (4,2,4)  и  (12,3,4)  и угол между ними.

__Определение__

Линейное пространтсво $R$ называется <i>нормированным</i>, если на нем введено правило, согласно которому каждому элементу $x$ ставится в соответствие число, называемое <i>нормой</i> (или <i>длиной</i>) этого элемента и обозначаемое символом $\left\|x\right\|$.

Оно должно подчиняться трем аксиомам:<br>
1) $\left\|x\right\|\geq0$, причем $\left\|x\right\|=0\Leftrightarrow x=0;$<br>
2) $\left\|\lambda x\right\|$ = $|\lambda|\left\|x\right\|$;<br>
3) $\left\| x + y \right\|$ <= $\left\|x\right\|$ + $\left\|y\right\|$.

Последняя аксиома называется <i>неравенством треугольника</i>, или <i>неравенством Минковского</i>.

__Примеры__

Существует много способов определния нормы.

- $p$-норма (норма Гёльдера) для вектора $x = (x_{1}, \dots, x_{n}) \in \mathbb{R}^{n}$ вычисляется по формуле:

$$\left\| x \right\|_{p} = \left( \sum_{i} \left| x_{i} \right|^{p} \right)^{1 / p},~p \geq 1.$$

В частных случаях при:

- $p = 1$ получаем $\ell_{1}$ норму (<i>манхэттенская норма</i>, или <i>манхэттенское расстояние</i>)

$$\left\|x\right\|_{1} = \sum_{i}|x_{i}|;$$

- $p = 2$ получаем $\ell_{2}$ норму (<i>евклидова норма</i>, наиболее часто используемая)

$$\left\|x\right\|_{2} = \sqrt{\sum_{i}|x_{i}|^{2}}.$$

```python
# Подключение библиотек
import numpy as np
from numpy.linalg import norm
```

```python
# Создание векторов
a = np.array([4, 2, 4])
b = np.array([12, 3, 4])
```

**Манхэттенская норма**

```python
norm(a, ord=1)
```

```python
norm(b, ord=1)
```

**Евклидова норма**

```python
norm(a, ord=2)
```

```python
norm(b, ord=2)
```
