#### Задание 1 
Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

$f_{3} = f_{1} + f_{2} + f_{4}.$ Вектор $f_{3}$ - линейная комбинация векторов $f_{1}(x)$, $f_{2}(x)$ и $f_{4}(x)$, из чего можно сделать вывод, что $f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$ линейно зависимы.

#### Задание 2
Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}.$$

Так как $f_{4}(x)=(x+1)^{2}=x^{2}+{2}x+{1}$ 

Тогда, $f_{4} = f_{3} + 2f_{2} + \frac{1}{2}f_{1}.$ Вектор $f_{4}$ - линейная комбинация векторов $f_{3}(x)$, $f_{2}(x)$ и $f_{1}(x)$, из чего можно сделать вывод, что $f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$ линейно зависимы.

#### Задание 3
Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

Воспользуемся методом подстановки для нахождения координат вектора:

$x=(\frac{1}{2}, 1, 3)=(\frac{1}{2},0,0)+(0, 1, 0)+(0, 0, 3)=\frac{1}{2}\cdot(0, 0, 10)+1\cdot(2, 0, 0)+3\cdot(0, 1, 0)=\frac{1}{2}b_{1}+1b_{2}+3b_{3},$

то есть, координатами вектора $x$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$ являются $(\frac{1}{2}$, $1$, $3)$.

Если выполнить проверку $\frac{1}{2}b_{1}+1b_{2}+3b_{3},$ то мы получим вектор $x$

#### Задание 4
Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

* В базисе $1$, $x$, $x^{2}$;

Методом подстановки найдём коодинаты вектора при базисе:
${3}\cdot(x^2)-{2}\cdot(x)+{2}\cdot(1) = 3x^{2}-2x+2$

Выпишим коэффициенты при базисах (в том порядке, в котором стоят базисы $1$, $x$, $x^{2}$.) - это будут координаты вектора

В данном базисе координаты вектора $(2, -2, 3)$

* В базисе $x^{2}$, $x-1$, $1$.

Методом подстановки найдём коодинаты вектора при базисе:
${3}\cdot(x^2)-{2}\cdot(x-1)+{0}\cdot(1) = 3x^{2}-2x+2$

Выпишим коэффициенты при базисах (в том порядке, в котором стоят базисы $x^{2}$, $x-1$, $1$.) - это будут координаты вектора

В данном базисе координаты вектора $(3, -2, 0)$
