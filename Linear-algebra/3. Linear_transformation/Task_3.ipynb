# Урок 3. Линейные преобразования



__4.__ Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
0 & 3 & 0\\ 
3 & 0 & 0\\
0 & 0 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(3, -3, -4)$ собственным вектором этого линейного оператора.

#### Задание 1
Найти собственные векторы и собственные значения для линейного оператора, заданного матрицей

$$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}.$$

**Вычислим характеристический многочлен для матрицы:**

$$det(\textbf{A}-\lambda\textbf{I})=\begin{vmatrix}
-1-\lambda & -6\\ 
2 & 6-\lambda
\end{vmatrix} = \lambda^{2}-(-1 + 6)\lambda+(-6 + 12)= \lambda^{2} - 5\lambda + 6$$

**Найдём собственные вектора:**

Вычислим корни уравнения $\lambda^{2} - 5\lambda + 6$

$x_1 = 2, x_2 = 3$

Теперь найдем собственные векторы вида $\begin{pmatrix}
x_{1}\\ 
x_{2}
\end{pmatrix}$, подставив полученное собственное значение в выражение $\textbf{A}x=\lambda x$:

__При $\lambda = 2$__


$$\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix} \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} = \lambda \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix}$$

Получим:

$$\begin{cases}
-x_{1}-6x_{2}= 2x_{1}, \\
2x_{1}+6x_{2}= 2x_{2}.
\end{cases}$$

Перенеся все в левую часть, получим:

$$\begin{cases}
-3x_{1}-6x_{2}= 0, \\
2x_{1}+4x_{2}= 0.
\end{cases}$$

Так как тут подобные строки, складываем их и получаем уравнение вида:

$-x_1 -2x_2 = 0$ Умножаем на -1 и переносим слагаемое:
$x_1 = -2x_2$

Теперь выполняем подстановку $x_1 = 1$, тогда $x_2 = -2$ и получаем вектор, вида $\begin{pmatrix}
-2\\ 
1\end{pmatrix}$

__Ответ:__ Множество векторов, коллинеарных вектору (-2, 1) - собственные.

__При $\lambda = 3$__


$$\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix} \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} = \lambda \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix}$$

Получим:

$$\begin{cases}
-x_{1}-6x_{2}= 3x_{1}, \\
2x_{1}+6x_{2}= 3x_{2}.
\end{cases}$$

Перенеся все в левую часть, получим:

$$\begin{cases}
-4x_{1}-6x_{2}= 0, \\
2x_{1}+3x_{2}= 0.
\end{cases}$$

Так как тут подобные строки, складываем их и получаем уравнение вида:

$-2x_1 -3x_2 = 0$ Умножаем на -1 и переносим слагаемое:
$2x_1 = -3x_2$

$x_1 = -\frac{1}{2}x_2$

Теперь выполняем подстановку $x_1 = 1$, тогда $x_2 = -0.5$ и получаем вектор, вида $\begin{pmatrix}
1\\ 
-0.5\end{pmatrix}$

__Ответ:__ Множество векторов, коллинеарных вектору (1, -0.5) - собственные.

#### Задание 2
Дан оператор поворота на 180 градусов, задаваемый матрицей 

$$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$$

Показать, что __любой__ вектор является для него собственным.

Собственное значение вектора ялвяется -1. Значит:

$$\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix} \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} = -1 \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix}$$

__Ответ:__ Данные выражения равны, тогда равенсто верно при любых х. Значит, любой вектор является собственным.
