### Задача № 1
`Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.`

Поскольку здесь важен не порядок, а лишь содержание, воспользуемся формулой сочетаний:

Изначально найдём все возможные сочетания:

$$C_{52}^4 = \frac{52!}{4! \: (52 - 4)!} = \frac{52!}{4! \: 48!} = \frac{49 \cdot 50 \cdot 51 \cdot 52}{4!} = \frac{49 \cdot 50 \cdot 51 \cdot 52}{2 \cdot 3 \cdot 4} = 270725$$

Найдём все возможные сочетания, где тузов нет (52 - 4)

$$C_{48}^4 = \frac{48!}{4! \: (48 - 4)!} = \frac{48!}{4! \: 44!} = \frac{45 \cdot 46 \cdot 47 \cdot 48}{4!} = \frac{45 \cdot 46 \cdot 47 \cdot 48}{2 \cdot 3 \cdot 4} = 194580$$

Итого получаем 270725 возможных сочетаний, которыми можно вытащить 4 карты и 194580 где тузов нет.
Вычитаем из всех сочетаний, те сочетания где тузов нет.

**Ответ: 270725 - 194580 = 76145**

### Задача № 2

`Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что рядом будут сидеть: а) два определённых человека? б) три определённых человека?`

**A)** Всего рассадок может быть 7! 
(A из n по k)

Человек может занять любое из 6 мест, каждая пара может сесть двумя способами (из-за этого число комбинаций требуется увеличить в два раза) и оставшихся можно рассажить 5! (так же по формуле размещений без повторений)

$$P(A) = \frac{6 \cdot 2 \cdot 5!}{7!} = \dfrac{2}{7}$$

**B)** Всего рассадок может быть 7! 
(A из n по k)

Человек может занять любое из 5 мест, далее по аналогии с прошлым получаем не 2, а 3! и рассаживаем остальных это 4!

$$P(A) = \frac{5 \cdot 3! \cdot 4!}{7!} = \dfrac{1}{7}$$

### Задача 3

`Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает: а) все? б) два?`

**A)** Найдём все способы, которыми можно достать 3 билета из 60 (сочетания без повторений):

$$C_{60}^3 = \frac{60!}{3! \: (60 - 3)!} = \frac{60 \cdot 59 \cdot 58}{3!} = 34220$$

Найдём число благоприятных исходов для вычисления вероятности:

$$C_{50}^3 = \frac{50!}{3! \: (50 - 3)!} = \frac{50 \cdot 49 \cdot 48}{3!} = 19600$$

Теперь можно определить вероятность для всех билетов:

$$P(A) = \dfrac{19600}{34220} = \dfrac{980}{1711} = 0,572764$$

**Б)** Все возможные исходы нам уже известны, осталось найти благоприятные исходы для вдух билетов (Выше делал подобное, только дял всех билетов, а именно для 3):

$$C_{50}^2 = \frac{50!}{2! \: (50 - 2)!} = \frac{50 \cdot 49}{2!} = 2450$$

Итоговое число получилось меньше, если посчитать вероятность, то я получу её ниже, а должно быт наоборот. Ведь веродяность вытянуть 2 билета выше, чем 3, значит где-то ошибка.

### Задача 4

`бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?`

Веротность выпадения чётного числа составляет и выпадения числа больше 3 тоже: 
$$P(A) = P(B) = \frac{1}{2}$$
Попробуем определить зависимые ли два событи друг отностительно друга:
Если мы знаем что что-то больше 3, тогда вероятность выпадения чётного числа составляет 2/3, а это значит, что события являются зависимыми.

### Задача 5

`Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.`

Вопрос: какова вероятность, что вы действительно больны ей?

Подсказка: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

Будем считать, что событие A - это пациент болен, событие B - тест положителен.

$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$

Используем формулу полной вероятности:
$$P(B) = P(B|A) \cdot P(A) + P(B|¬A) \cdot P(¬A)$$

$$P(B) = 0.99 \cdot 0,001 + 0.01 \cdot 0.999 = 0.01098$$
